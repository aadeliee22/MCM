```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cols = ['m', 'ms', 'e', 'sp_h']
lbl = ['eff', 'srch', 'class']

def plot(c, m):
    for j in range (4):
        plt.plot(m.temperature, m[cols[j]], 'r.', markersize = 10, markerfacecolor = None, linestyle = 'None', label = 'original')
        for i in range (len(c)):
            plt.plot(c[i].temperature, c[i][cols[j]], '.', markersize = 3, markerfacecolor = None, label = lbl[i])
        plt.ylabel(cols[j], fontsize=16)
        plt.xlabel("T", fontsize=16)
        plt.legend()
        plt.show()
def plot2(c,m):
    for i in range (2):
        R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
        print(lbl[i], ":", R2)
        plt.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('cluster', fontsize=12)
    plt.ylabel('original', fontsize=12)
    plt.legend()
    plt.show()
```

```python
# J2: 0.2 ~ -0.2
J2 = -0.05

path = '../txtfiles-5/J2_pm0.2'
cc = []
cc.append(pd.read_csv(path + '/cluster_eff_2.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_2.txt', sep = ' ', header = 0))
#cc.append(pd.read_csv(path + '/class_test.txt', sep= ' ', header = 0))
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]

print("J2 =", np.average(m.J2))
plot(c, m)
```

```python
# J2: 0.1 ~ -0.1  # J3: 0.075 ~ -0.075

J2 = -0.05
J3 = -0.05

path = '../txtfiles-5/J3_pm0.075'
cc = []

cc.append(pd.read_csv(path + '/cluster_eff_3.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_3.txt', sep = ' ', header = 0))
mm = pd.read_csv(path + '/met_3.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2][cc[i][cc[i].J2==J2].J3==J3])
m = mm[mm.J2==J2][mm[mm.J2==J2].J3==J3]

print("J2 =", np.average(m.J2), " & J3 =", np.average(m.J3))
plot2(c,m)
```
