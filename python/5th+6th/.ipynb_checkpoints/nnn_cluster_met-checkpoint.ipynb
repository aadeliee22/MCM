```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

cols = ['m', 'ms', 'e', 'sp_h']
lbl = ['2.2', '2.1']

def plot(c, m):
    for j in range (4):
        plt.plot(m.temperature, m[cols[j]], 'r.', markersize = 10, markerfacecolor = None, linestyle = 'None', label = 'original')
        for i in range (len(c)):
            plt.plot(c[i].temperature, c[i][cols[j]], '.', markersize = 3, markerfacecolor = None, label = lbl[i])
        plt.ylabel(cols[j], fontsize=16)
        plt.xlabel("T", fontsize=16)
        plt.legend()
        plt.show()
def plot2(c,m):
    for i in range (2):
        R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
        print(lbl[i], ":", R2)
        plt.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.xlabel('cluster', fontsize=12)
    plt.ylabel('original', fontsize=12)
    plt.legend()
    plt.show()
```

```python
# J2: 0.2 ~ -0.2
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize=(15,3))

J2 = -0.15
path = '../txtfiles-5/J2_pm0.2'
cc = []
cc.append(pd.read_csv(path + '/cluster_eff_2.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_2.txt', sep = ' ', header = 0))
#cc.append(pd.read_csv(path + '/class_test.txt', sep= ' ', header = 0))
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]
for i in range (2):
    R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
    print(lbl[i], ":", R2)
    ax1.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    ax1.set(title = J2)
    plt.gca().set_aspect('equal', adjustable='box')

J2 = -0.05
path = '../txtfiles-5/J2_pm0.2'
cc = []
cc.append(pd.read_csv(path + '/cluster_eff_2.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_2.txt', sep = ' ', header = 0))
#cc.append(pd.read_csv(path + '/class_test.txt', sep= ' ', header = 0))
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]
for i in range (2):
    R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
    print(lbl[i], ":", R2)
    ax2.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    ax2.set(title = J2)
    plt.gca().set_aspect('equal', adjustable='box')
    
J2 = 0.05
path = '../txtfiles-5/J2_pm0.2'
cc = []
cc.append(pd.read_csv(path + '/cluster_eff_2.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_2.txt', sep = ' ', header = 0))
#cc.append(pd.read_csv(path + '/class_test.txt', sep= ' ', header = 0))
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]
for i in range (2):
    R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
    print(lbl[i], ":", R2)
    ax3.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    ax3.set(title = J2)
    plt.gca().set_aspect('equal', adjustable='box')
    
J2 = 0.15
path = '../txtfiles-5/J2_pm0.2'
cc = []
cc.append(pd.read_csv(path + '/cluster_eff_2.txt', sep = ' ', header = 0))
cc.append(pd.read_csv(path + '/cluster_search_2.txt', sep = ' ', header = 0))
#cc.append(pd.read_csv(path + '/class_test.txt', sep= ' ', header = 0))
mm = pd.read_csv(path + '/met_2.txt', sep = ' ', header = 0)
c = []
for i in range (len(cc)):
    c.append(cc[i][cc[i].J2==J2])
m = mm[mm.J2==J2]
for i in range (2):
    R2 = 1 - np.sum((np.array(c[i].e) - np.array(m.e))**2) / np.sum((c[i].e - np.average(c[i].e))**2)
    print(lbl[i], ":", R2)
    ax4.plot(c[i].e, m.e, '.', markersize = 5, label = lbl[i])
    ax4.set(title = J2)
    plt.gca().set_aspect('equal', adjustable='box')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax1.set_xlabel('cluster', fontsize=12)
ax2.set_xlabel('cluster', fontsize=12)
ax3.set_xlabel('cluster', fontsize=12)
ax4.set_xlabel('cluster', fontsize=12)
ax1.set_ylabel('original', fontsize=12)    
plt.show()
```
