```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
c = []
c.append(pd.read_csv('cluster_1.txt', sep = ' ', header = 0)) #1
c.append(pd.read_csv('cluster_2.txt', sep = ' ', header = 0)) #2
m = pd.read_csv('met_2.txt', sep = ' ', header = 0)

cols = ['m', 'ms', 'e', 'sp_h']
```

```python

print("J2 =", m.J2[0])

for j in range (4):
    plt.plot(m.temperature, m[cols[j]], 'r.', markersize = 10, markerfacecolor = None, linestyle = 'None', label = 'original')
    for i in range (2):
        plt.plot(c[i].temperature, c[i][cols[j]], '.', markersize = 3, markerfacecolor = None, label = i+1)
    plt.ylabel(cols[j], fontsize=16)
    plt.xlabel("T", fontsize=16)
    plt.legend()
    plt.show()
```
