```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = []
p.append(pd.read_csv('p10.txt', sep=' ', header=0))
p.append(pd.read_csv('p20.txt', sep=' ', header=0))
p.append(pd.read_csv('p40.txt', sep=' ', header=0))
cases=len(p)
a10 = []
a10.append(pd.read_csv('paper_10_1.txt', sep = ' ', header = 0)) #1
a10.append(pd.read_csv('paper_10_2.txt', sep = ' ', header = 0)) #2
a10.append(pd.read_csv('paper_10_3.txt', sep = ' ', header = 0)) #3
a20 = []
#a20.append(pd.read_csv('paper_20_1.txt', sep = ' ', header = 0)) #1
#a20.append(pd.read_csv('paper_20_2.txt', sep = ' ', header = 0)) #2
#a20.append(pd.read_csv('paper_20_3.txt', sep = ' ', header = 0)) #3
ap = []
ap.append(pd.read_csv('paper_10_1_p.txt', sep = ' ', header = 0)) #1
ap.append(pd.read_csv('paper_10_3_p.txt', sep = ' ', header = 0)) #3
```

```python
print("Plaquette of L = 10, 20, 40")
for i in range (cases):
    plt.plot(p[i].temperature, p[i].m, 'o', markersize=1, label=2**i*10)
plt.xlabel('T', fontsize=14)
plt.ylabel('$<m>$', fontsize=14)
plt.xlim(1.5, 3.5)
plt.ylim(0)
plt.legend()
plt.show()
for i in range (cases):
    plt.plot(p[i].temperature, p[i].ms, 'o', markersize=1, label=2**i*10)
plt.xlabel('T', fontsize=14)
plt.ylabel('$\chi$', fontsize=14)
plt.xlim(1.5, 3.5)
plt.legend()
plt.show()
for i in range (cases):
    plt.plot(p[i].temperature, p[i].e/(2**i*10)**2, 'o', markersize=1, label=2**i*10)
plt.xlabel('T', fontsize=14)
plt.ylabel('E/N', fontsize=14)
plt.xlim(1.5, 3.5)
plt.legend()
plt.show()
for i in range (cases):
    plt.plot(p[i].temperature, p[i].sp_h, 'o', markersize=1, label=2**i*10)
plt.xlabel('T', fontsize=14)
plt.ylabel('C', fontsize=14)
plt.xlim(1.5, 3.5)
plt.legend()
plt.show()

T_c = 2.493
for i in range (cases):
    plt.plot(p[i].temperature, 1-p[i].m4/(3*p[i].m2**2), '.', markersize=1, label=2**i*10)
plt.axhline(y=0, color='k', linewidth=1, linestyle='--')
plt.axvline(x=T_c, color='k', linewidth=1, linestyle='--')
plt.xlim(1.5, 3.5)
plt.xlabel('T', fontsize=14)
plt.ylabel('$B_L$', fontsize=14)
plt.legend()
plt.show()
```
