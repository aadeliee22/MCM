```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = []
ppath = '/home/hyejin/바탕화면/mygit/MCM/python/txtfiles-5/'
p.append(pd.read_csv(ppath + 'p10.txt', sep=' ', header=0))
p.append(pd.read_csv(ppath + 'p20.txt', sep=' ', header=0))
p.append(pd.read_csv(ppath + 'p40.txt', sep=' ', header=0))
cases=len(p)
a10 = []
#a10.append(pd.read_csv('paper_10_1.txt', sep = ' ', header = 0)) #1
#a10.append(pd.read_csv('paper_10_2.txt', sep = ' ', header = 0)) #2
#a10.append(pd.read_csv('paper_10_3.txt', sep = ' ', header = 0)) #3
a20 = []
#a20.append(pd.read_csv('paper_20_1.txt', sep = ' ', header = 0)) #1
#a20.append(pd.read_csv('paper_20_2.txt', sep = ' ', header = 0)) #2
#a20.append(pd.read_csv('paper_20_3.txt', sep = ' ', header = 0)) #3
ap = []
#ap.append(pd.read_csv('inpaper_10_1.txt', sep = ' ', header = 0)) #1
#ap.append(pd.read_csv('inpaper_10_3.txt', sep = ' ', header = 0)) #3

cols = ['m', 'ms', 'e', 'sp_h']
```

```python
print("Plaquette of L = 10, 20, 40")
for j in range (4):
    for i in range (cases):
        plt.plot(p[i].temperature, p[i][cols[j]], 'o', markersize=1, label=2**i*10)
    plt.xlabel('T', fontsize=14)
    plt.ylabel(cols[j], fontsize=14)
    plt.xlim(1.5, 3.5)
    plt.legend()
    plt.show()

T_c = 2.493
for i in range (cases):
    plt.plot(p[i].temperature, 1-p[i].m4/(3*p[i].m2**2), '.', markersize=1, label=2**i*10)
plt.axhline(y=0, color='k', linewidth=1, linestyle='--')
plt.axvline(x=T_c, color='k', linewidth=1, linestyle='--')
plt.xlim(1.5, 3.5)
plt.xlabel('T', fontsize=14)
plt.ylabel('$B_L$', fontsize=14)
plt.legend()
plt.show()
```

```python
print("Plaquette & Fitted of L=10")

for j in range (4):
    plt.plot(p[0].temperature, p[0][cols[j]], 'ro', markersize=6, label = 10, markerfacecolor = 'None', linestyle = 'None')
    for i in range (len(a10)):
        plt.plot(a10[i].temperature, a10[i][cols[j]], 'o', markersize=1, label=i+1)
    plt.xlabel('T', fontsize=14)
    plt.ylabel(cols[j], fontsize=14)
    plt.xlim(1.5, 3.5)
    plt.legend()
    plt.show()

T_c = 2.493
plt.plot(p[0].temperature, 1-p[0].m4/(3*p[0].m2**2), 'ro', markersize=1, label=10*(1+i))
for i in range (len(a10)):
    plt.plot(a10[i].temperature, 1-a10[i].m4/(3*a10[i].m2**2), 'o', markersize=1, label=i+1)
plt.axhline(y=0, color='k', linewidth=1, linestyle='--')
plt.axvline(x=T_c, color='k', linewidth=1, linestyle='--')
plt.xlim(1.5, 3.5)
plt.xlabel('T', fontsize=14)
plt.ylabel('$B_L$', fontsize=14)
plt.legend()
plt.show()
```
