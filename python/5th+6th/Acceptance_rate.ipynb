```python
import numpy as np
import pandas as pd
import subprocess as sub
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
p = []
p.append(pd.read_csv('../txtfiles-6/acceptance/8_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/8_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/16_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/16_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/32_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/32_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/48_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/48_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/64_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/64_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/96_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/96_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/128_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/128_3.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/256_1.txt', sep = ' ', header = 0))
p.append(pd.read_csv('../txtfiles-6/acceptance/256_3.txt', sep = ' ', header = 0))
num = 8
```

```python
ac = np.zeros((num, 2, 3, 3))
for i in range (num):
    for j in range (2):
        for k in range (3):
            ac[i][j][k][0] = p[2*i+j].ac1[k]/p[2*i+j].step2[k]
            ac[i][j][k][1] = p[2*i+j].aca[k]/p[2*i+j].ac1[k]
            ac[i][j][k][2] = p[2*i+j].aca[k]/p[2*i+j].step2[k]
```

```python
xx = np.array([8, 16, 32, 48, 64, 96, 128, 256])
lbs = ['1', '3']
clr = ['red', 'blue']
lines = [Line2D([0], [0], color='r', marker = 'o', lw=1),
                Line2D([0], [0], color='b', marker = 'o', lw=1)]
fig, ax = plt.subplots(1, 3, figsize = (17, 4))
for l in range (3):
    for j in range (2):
        ax[l].plot(xx, ac[:, j, :, l], 'o-', color = clr[j], label = lbs[j])
        ax[l].set_xlabel('L', fontsize = 15)
        ax[l].set_ylim(0, 1.05)
        ax[l].set_xscale('log')
ax[0].set_ylabel('$\\alpha$', fontsize=15)
ax[0].set_title('Acceptance lv=1', fontsize= 14)
ax[1].set_title('Acceptance lv=2', fontsize= 14)
ax[2].set_title('Total Acceptance', fontsize= 14)
ax[2].legend(lines, ['1', '3'], loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
