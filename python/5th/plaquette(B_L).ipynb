```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = []
p.append(pd.read_csv('plaquette_m1.txt', sep=' ', header=0))
p.append(pd.read_csv('plaquette_m2.txt', sep=' ', header=0))
cases=len(p)
q = pd.read_csv('10_checker_1.txt', sep=' ', header=0)
```

```python
for i in range (cases):
    plt.plot(p[i].temperature, p[i].m, '.', markersize=1, label=10*(1+i))
plt.plot(q.temperature, q.m, '.', markersize=1)
plt.xlabel('T', fontsize=14)
plt.ylabel('$<m>$', fontsize=14)
plt.xlim(1.5, 3.5)
plt.ylim(0)
plt.legend()
plt.show()
```

```python
for i in range (cases):
    plt.plot(p[i].temperature, p[i].mag_sus, '.', markersize=1, label=10*(1+i))
plt.plot(q.temperature, q.mag_sus)
plt.xlabel('T', fontsize=14)
plt.ylabel('$\chi$', fontsize=14)
plt.xlim(1.5, 3.5)
plt.legend()
plt.show()
```
