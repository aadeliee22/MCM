```python
T_c=2/np.log(1+np.sqrt(2)) #critical temperature
nu = 1
gamma = 7/4
plt.plot((f16.temperature-T_c)*16**(1/nu), f16.mag_sus*16**(-gamma/nu), 'b.', markersize=1, label='L=16')
plt.plot((f32.temperature-T_c)*32**(1/nu), f32.mag_sus*32**(-gamma/nu), 'g.', markersize=1, label='L=32')
plt.plot((f48.temperature-T_c)*48**(1/nu), f48.mag_sus*48**(-gamma/nu), 'r.', markersize=1, label='L=48')
plt.plot((f64.temperature-T_c)*64**(1/nu), f64.mag_sus*64**(-gamma/nu), 'y.', markersize=1, label='L=64')
a16 = ((f16.temperature-T_c)*16**(1/nu))[np.argmax(f16.mag_sus*16**(-gamma/nu))]
a32 = ((f32.temperature-T_c)*32**(1/nu))[np.argmax(f32.mag_sus*32**(-gamma/nu))]
a48 = ((f48.temperature-T_c)*48**(1/nu))[np.argmax(f48.mag_sus*48**(-gamma/nu))]
a64 = ((f64.temperature-T_c)*64**(1/nu))[np.argmax(f64.mag_sus*64**(-gamma/nu))]
a = (a16+a32+a48+a64)/4
print("maximum location =", a)
plt.xlim(-30, 30)
plt.xlabel('(T-T_c)$\cdot$L^(1/$ \\nu$)', fontsize=14)
plt.ylabel('$\chi\cdot$L^(-$\gamma$/$\\nu$)', fontsize=14)
plt.axvline(x=0, color='k', linewidth=1)
plt.axvline(x=a, color='r', linewidth=1)
plt.legend()
plt.show()
```

```python
max_16=np.zeros(5)
max_32=np.zeros(5)
max_48=np.zeros(5)
max_64=np.zeros(5)

for i in range (5):
    max_16[i]=f16.temperature[f16.mag_sus==sorted(f16.mag_sus, reverse=True)[i]]
    max_32[i]=f32.temperature[f32.mag_sus==sorted(f32.mag_sus, reverse=True)[i]]
    max_48[i]=f48.temperature[f48.mag_sus==sorted(f48.mag_sus, reverse=True)[i]]
    max_64[i]=f64.temperature[f64.mag_sus==sorted(f64.mag_sus, reverse=True)[i]]
pseudo16=np.average(max_16)
pseudo32=np.average(max_32)
pseudo48=np.average(max_48)
pseudo64=np.average(max_64)
```
